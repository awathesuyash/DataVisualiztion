```python
#Compute the x and y coordinate for points on a sine curve

x=np.arange(0,4 *np.pi,0.1)
y=np.sin(x)

plt.title('Sine waveform')


#Plot the points uing matplot lib

plt.plot(x,y,color='r')
plt.show()
```

```python
x=np.arange(1,10)
y= 3*x +10

#plot
plt.plot(x,y,'r3--')
plt.xlabel('x-axis',color='blue')
plt.ylabel('y-axis',color='purple')
plt.title('2d-diagram',color='green')
```

```python
#creating Subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.show()
```
